```python
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])

# Extract Date and Time separately
df['hour'] = df['trans_date_trans_time'].dt.hour
```

```python
features = df.drop('is_fraud', axis=1)
features = features.dropna()
```

```python
target=df['is_fraud']
target = target.dropna()
```

```python
features = features.drop(['Unnamed: 0','first','last','city','dob','gender','cc_num','trans_num','unix_time','street','job','trans_date_trans_time','state','merchant' ],axis=1)
```

```python
features = pd.get_dummies(features, columns=['category'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.1)
scalar = StandardScaler()

X_train = scalar.fit_transform(X_train)
X_test = scalar.fit_transform(X_test)
```

```python
def evaluate_and_save_model(model, X_train, X_test, y_train, y_test, filename):
  model.fit(X_train, y_train)
  y_pred= model.predict(X_test)
  accuracy = accuracy_score(y_test,y_pred)
  print(f"{model.__class__.__name__} Accuracy: {accuracy:.4f}")
  print(f"\nClassification Report: \n{classification_report(y_test,y_pred)}")
  print("------------")

  with open(filename, 'wb') as file:
    pickle.dump(model,file)

  print(f"Model saved as {filename}")
```

```python
xgb_model = xgb.XGBClassifier(random_state=40, learning_rate=0.7, min_split_loss=3, max_depth=10)
evaluate_and_save_model(xgb_model, X_train, X_test, y_train, y_test, "xgb_model1.pkl")

dt_model = DecisionTreeClassifier(random_state=40)
evaluate_and_save_model(dt_model, X_train, X_test, y_train, y_test, "dt_model1.pkl")

#rf_model = RandomForestClassifier(random_state=40)
#evaluate_and_save_model(rf_model, X_train, X_test, y_train, y_test, "rf_model.pkl")

mb_model = GaussianNB()
evaluate_and_save_model(mb_model, X_train, X_test, y_train, y_test, "mb_model.pkl")

#knn_model = KNeighborsClassifier()
#evaluate_and_save_model(knn_model, X_train, X_test, y_train, y_test, "knn_model.pkl")

#svm_model = SVC(random_state=40)
#evaluate_and_save_model(svm_model, X_train, X_test, y_train, y_test, "svm_model.pkl")

from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(class_weight='balanced', random_state=40)
evaluate_and_save_model(lr_model, X_train, X_test, y_train, y_test, "lr_model.pkl")

from sklearn.ensemble import HistGradientBoostingClassifier

hgb_model = HistGradientBoostingClassifier(random_state=40)
evaluate_and_save_model(hgb_model, X_train, X_test, y_train, y_test, "hgb_model1.pkl")
```

```python
def print_feature_importance(model, model_name, feature_names):
    try:
        if hasattr(model, 'feature_importances_'):
            print(f"{model_name} Feature Importances:")
            importances = model.feature_importances_
            sorted_idx = np.argsort(importances)[::-1]  # Sort by importance
            for idx in sorted_idx:
                print(f"{feature_names[idx]}: {importances[idx]:.4f}")
        elif hasattr(model, 'coef_'):
            print(f"{model_name} Coefficients:")
            for feature, coef in zip(feature_names, model.coef_[0]):
                print(f"{feature}: {coef:.4f}")
        else:
            print(f"{model_name} does not support feature importance.")
    except Exception as e:
        print(f"Error retrieving feature importance for {model_name}: {e}")
print_feature_importance(xgb_model, "XGBoost", feature_names)
print_feature_importance(dt_model, "Decision Tree", feature_names)
print_feature_importance(lr_model, "Logistic Regression", feature_names)
print_feature_importance(hgb_model, "HistGradientBoosting", feature_names)
```
